```python
import numpy as np
from tqdm import tqdm
import scipy.sparse as sp
import scanpy as sc

def read_adata(h5ad_path):
    
    
    adata = sc.read_h5ad(h5ad_path)
    return adata
```

```python
def get_col_dict(
    adata,
    cols=("plate", "cell_line", "drugname_drugconc"),
):

    # 1) obs三列一次性取出，避免每行反复做 pandas 索引
    col_dict = {}
    for c in cols:
        col_dict[c] = adata.obs[c].unique()
    return col_dict


# 关注的cols
cols_needed = ("perturbation",)
global_keys = {}
for k in cols_needed:
    global_keys[k] = []


h5ad_path = f"./only_hvg/Norman_only_hvg.h5ad"
adata = read_adata(h5ad_path)
adata_col_dict = get_col_dict(adata, cols_needed)

for col_name, col_list in adata_col_dict.items():
    global_keys[col_name].extend(col_list)
    
    
for k in global_keys.keys():
    global_keys[k] = sorted(list(set(global_keys[k])))


global_keys
```

```python
test = [x for x in global_keys['perturbation'] if '_' not in x]
single_perturbation = []
for x in global_keys['perturbation']:
    perturbation_list = []
    if '_' in x:
        perturbation_list = x.split('_')
    else:
        perturbation_list.append(x)
    single_perturbation.extend(perturbation_list)

single_perturbation = list(set(single_perturbation))
print(len(test))
print(len(single_perturbation))


```
