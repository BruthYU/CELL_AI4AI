```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm

folder = "./obs_dataframe"   # 改成你的目录
plate_ids = range(1,15)

all_drugname_drugconc = set()
all_drug = set()
all_cell_line = set()

for pid in tqdm(plate_ids):
    fp = os.path.join(folder, f"plate{pid}_obs.csv")
    if not os.path.exists(fp):
        continue

    df = pd.read_csv(fp, sep="\t")

    u1 = pd.Series(df["drugname_drugconc"].dropna().unique()).astype(str).str.strip()
    u2 = pd.Series(df["drug"].dropna().unique()).astype(str).str.strip()
    u3 = pd.Series(df["cell_line"].dropna().unique()).astype(str).str.strip()

    all_drugname_drugconc.update(u1.tolist())
    all_drug.update(u2.tolist())
    all_cell_line.update(u3.tolist())

# 转成 list（排序是为了可复现；不需要排序就去掉 sorted）
drugname_drugconc_list = sorted(all_drugname_drugconc)
drug_list = sorted(all_drug)
cell_line_list = sorted(all_cell_line)

# 保存为 npy（object array）
np.save(os.path.join(folder, "drugname_drugconc_union.npy"),
        np.array(drugname_drugconc_list, dtype=object))
np.save(os.path.join(folder, "drug_union.npy"),
        np.array(drug_list, dtype=object))
np.save(os.path.join(folder, "cell_line_union.npy"),
        np.array(cell_line_list, dtype=object))

print("saved:",
      os.path.join(folder, "drugname_drugconc_union.npy"),
      os.path.join(folder, "drug_union.npy"),
      os.path.join(folder, "cell_line_union.npy"))
      
```
