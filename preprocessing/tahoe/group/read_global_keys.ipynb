```python
import numpy as np
from tqdm import tqdm
import scipy.sparse as sp
import scanpy as sc

def read_adata(plate_num=3):
    plate=f'plate{plate_num}'
    raw_h5ad_path = f"/mnt/shared-storage-gpfs2/beam-gpfs02/yulang/datasets/by_plate/hvg/{plate}_hvg.h5ad"
    adata = sc.read_h5ad(raw_h5ad_path)
    return adata
```

```python
def get_col_dict(
    adata,
    cols=("plate", "cell_line", "drugname_drugconc"),
):

    # 1) obs三列一次性取出，避免每行反复做 pandas 索引
    col_dict = {}
    for c in cols:
        col_dict[c] = adata.obs[c].unique()
    return col_dict


# 关注的cols
cols_needed = ("plate", "cell_line", "drugname_drugconc")
global_keys = {}
for k in cols_needed:
    global_keys[k] = []

for i in range(1, 15):
    print(f"Processing plate {i}")
    adata = read_adata(i)
    adata_col_dict = get_col_dict(adata, cols_needed)

    for col_name, col_list in adata_col_dict.items():
        global_keys[col_name].extend(col_list)
    
    del adata
for k in global_keys.keys():
    global_keys[k] = sorted(list(set(global_keys[k])))
```

```python
for k in global_keys.keys():
    print(k, len(global_keys[k]))
```

```python
import pickle
with open("global_keys.pkl", "wb") as f:
    pickle.dump(global_keys, f)
```
